```python
import numpy as np
import keras
```

```python
model = keras.Sequential([keras.layers.Dense(units =1, input_shape = [1])])
```

```python
model.compile(optimizer="sgd", loss = "mean_squared_error")
```

```python
x = np.array([-1, 0, 1,2,3], dtype = float)
```

```python
y = np.array([-3, -1, 1, 3, 5], dtype = float)
# relationship is y = 2x-1
```

```python
model.fit(x,y,epochs = 10)
# 10 epochs
```

```python
model.predict([110, 10, 0])
# predicted o/p so for from actual
```

```python
model.fit(x,y,epochs = 100)
# with 100 expochs
```

```python
model.predict([110, 10, 0])
#given better almost 90+%
```

```python
model.fit(x,y,epochs = 500)
```

```python
model.predict([110, 10, 0])
#given better almost 98+%
```
