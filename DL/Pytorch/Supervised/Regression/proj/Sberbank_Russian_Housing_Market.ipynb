```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/Kaggle sber bank/processed_df.csv')
```

```python
import torch
```

```python
df.describe()
```

```python
Features = df["full_sq"]
labels = df.iloc[:,-1:]
```

```python
X_Train = torch.FloatTensor(Features.values)
y_train = torch.FloatTensor(labels.values)
```

```python
X_Train
```

```python
y_train
```

```python
X_Train.shape
```

```python
linear = torch.nn.Linear(1,1)
```

```python
loss_fun = torch.nn.MSELoss()
```

```python
learning_rate = 0.001
optimizer = torch.optim.SGD(linear.parameters(), lr=learning_rate)
```

```python
for epoch in range(200):
    y_pred = linear(X_Train)
    loss = loss_fun(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
        
    if (epoch+1) % 5 == 0:
        print ('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, 200, loss.item()))
```
