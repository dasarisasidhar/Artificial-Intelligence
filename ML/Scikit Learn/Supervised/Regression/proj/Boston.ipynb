```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
train_df = pd.read_csv('../Datasets/Boston/train.csv')
test_df = pd.read_csv('../Datasets/Boston/test.csv')
```

```python
train_df.head()
```

```python
test_df['SalePrice'] = -1234
```

```python
#feature engineering process1 
#Remove all the null and missing values and convert string to numerical and continue
#finding all the missing values in rows and remove

train_df = train_df.dropna(axis=1,how='all')
df = pd.concat((train_df, test_df), axis = 0)
```

```python
df
```

```python
#feature engineering process1 
#Remove all the null and missing values and convert string to numerical and continue
#finding all the missing cols
df.columns[df.isnull().any()]
```

```python
df[df.SalePrice != -1234].columns.isnull().any()
```

```python
df.loc[:, df.dtypes == object].count()
```

```python
for i in df.loc[:, df.dtypes == object]:
    print(df[i].unique())
    freq_map = df[i].value_counts().to_dict()
    print(freq_map)
    df[i] = df[i].map(freq_map)
```

```python
df.drop(["Alley", "PoolQC", "Fence", "MiscFeature"], inplace = True, axis = 1)
```
