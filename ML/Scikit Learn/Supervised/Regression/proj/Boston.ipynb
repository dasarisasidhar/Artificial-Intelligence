```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
train_df = pd.read_csv('../Datasets/Boston/processed_train.csv')
```

```python
train_df.head()
```

```python
train_df.SalePrice
```

```python
Features = train_df.loc[:,train_df.columns != 'SalePrice']
```

```python
labels =  train_df.loc[:,'SalePrice']
```

```python
Features
```

```python
labels
```

```python
X_train, X_test, y_train, y_test = train_test_split(Features, labels, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
rid = Ridge(alpha=.5)
rid.fit(X_train, y_train)
rid.score(X_test,y_test)
```

```python
rid_predict = rid.predict(X_test)
```

```python
rid_predict.shape
```

```python
y_test.shape
```

```python
sgd_reg = SGDRegressor(max_iter=100, penalty=None, eta0=0.001)
sgd_reg.fit(X_test, y_test)
sgd_reg.score(X_test,y_test)
```

```python
sgd_reg.intercept_, sgd_reg.coef_
```

```python
lasso = Lasso(alpha = 0.1, normalize = True)
lasso.fit(X_train,y_train)
lasso_predict = lasso.predict(X_test)
lasso.score(X_test,y_test)
```
