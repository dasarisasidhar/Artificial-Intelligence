```python
import numpy as np
import pandas as pd
```

```python
train_data_frame = pd.read_csv('../Datasets/titanic/train.csv')
test_data_frame = pd.read_csv('../Datasets/titanic/test.csv')
```

```python
train_data_frame
```

```python
test_data_frame['Survived'] = -888  # adding Survived with default value to test df
```

```python
#add both train and test data
data = pd.concat((train_data_frame, test_data_frame), axis = 0) #axis is your to concat row wise
```

```python
data
```

```python
data = data[['PassengerId','Pclass', 'Sex', 'Age', 'Survived']]
```

```python
#labels = train_data_frame.Survived
```

```python
#labels
```

```python
data
```

```python
data['Sex'] = data['Sex'].map({'male': 0, 'female':1})
```

```python
test_data = data[data['Survived'] == -888]
data = data[data['Survived'] != -888]
```

```python
data = data.dropna()
```

```python
test_data
```

```python
data = data.drop("PassengerId", axis=1)
```

```python
X = data.drop("Survived", axis=1)
y = data["Survived"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from xgboost import XGBClassifier
```

```python
my_model = XGBClassifier()

my_model.fit(X_train, y_train, verbose=False)
```

```python
my_model.predict(X_test)
```

```python
my_model.score(X_train, y_train)
```
